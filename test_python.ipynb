Le fichier csv contient la température mais codée avec une virgule. Dans python, les nombres doivent contenir des points pour être valides. Voir exemple dans le fichier csv:
```
03/11/2020;00:00;chambrefroide01;2,5
```
Pour faire la modification, vous pouvez utiliser la fonction replace qui permet de remplacer un caractère par un autre.
Voir la documentation de cette fonction: https://www.tutorialspoint.com/python/string_replace.htm
Vous pouvez vous inspirer de l'exemple ci-dessous.
```
temperature_corrigee = champ[3].replace(',', '.')
temperature = float(temperature_corrigee)
```

````python
# Code source exercice 2
# affichage des temperatures v1
import csv

# Affichage de l'entete du programme
print("Affichage des mesures de température d'une chambre froide")
print('-'*57)
minimum = 1.0
maximum = 4.0

# Selection d'une chambre froide
num_chambrefroide = input('Choisissez un numéro de chambre froide (1, 2 ou 3) : ')

if num_chambrefroide in ('1','2','3'):
    # Nom de la chambre froide a rechercher dans les données
    recherche = 'chambrefroide0' + num_chambrefroide
    
    # Ouverture du fichier de données
    with open('traiteur.csv') as csv_file:
        contenu_csv = csv.reader(csv_file, delimiter=';')
        
        
        # parcours des données du fichier
        next(contenu_csv)
        for champ in contenu_csv:
            champ[3] = champ[3].replace(",", ".")
            if float(champ[3]) < minimum or float(champ[3]) > maximum:
                champ[3] += " **Alarme**"
            if champ[2] == recherche:
                print(champ[0],champ[1],champ[2],champ[3])
else:
    print("erreur: numéro de chambre froide incorrect")
````
